```python
from tkinter import *

window = Tk()
window.geometry('400x250')
window.title('Click event')

def callback(event):
    print('X, Y:', event.x, event.y)
    print('Widget:', event.widget)
    print('X-root, Y-root:', event.x_root, event.y_root)
    
def click(event):
    print('You click on me seccond time')
    
def enter(event):
    print('You enter me')
    event.widget['bg'] = 'blue'
    
def leave(event):
    print('You leave me')
    event.widget['bg'] = 'red'
    
def entered(event):
    print('Oh, you press enter on me')
    
def key(event):
    print('Key pressed: ', event.char)

    
button = Button(window, text='Confirm')
button2 = Button(window, text='Exit')
# frame = Frame(window, width=400, height=250)
button.bind('<Button-1>', callback)
button.bind('<Button-1>', click, add='+')
button.bind('<Return>', entered)
button.bind('<Key>', key)
# button.bind('<Enter>', enter)
# button.bind('<Leave>', leave)
button.focus()
button.pack()
button2.pack()

window.mainloop()
```

```python
from tkinter import *

window = Tk()
window.geometry('400x250')
window.title('Click event')

def gt(event):
    number = txtNumber.get()
    result = 1
    for i in range(2, int(number)+1):
        result *= i
    lbMessage['text'] = result
    
txtNumber = Entry(window, text='')
txtNumber.pack()

lbMessage = Label(window, text='Out Message')
lbMessage.pack()

button = Button(window, text='Giai thừa')
button.bind('<Button-1>', gt)
button.pack()

window.mainloop()
```

```python
from tkinter import *

def gt(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            result = 1
            for i in range(2, int(number)+1):
                result *= i
            labelOutput['text'] = result
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
    
def lt(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            number = entryInput.get()
            labelOutput['text'] = (int(number))**2
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
def cb2(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            number = entryInput.get()
            labelOutput['text'] = (int(number))**0.5
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
        
window = Tk()
window.geometry('400x250')
window.title('Click event')

titleInput = Label(window, text='Input:')
titleInput.place(x=10, y= 10)
titleOutput = Label(window, text='Output:')
titleOutput.place(x=10, y= 50)

entryInput = Entry(window, text = '', width = 35)
entryInput.place(x=100, y = 10)
labelOutput = Label(window, text = '', borderwidth =1)
labelOutput.place(x=100, y = 50)
entryInput.focus()

btGT = Button(window, text='Giai thừa')
btGT.bind('<Button-1>', gt)
btGT.place(x=30, y = 100, width=70, height=30)

btLT = Button(window, text='Lũy thừa')
btLT.bind('<Button-1>', lt)
btLT.place(x=120, y = 100, width=70, height=30)

btCB2 = Button(window, text='Căn bậc 2')
btCB2.bind('<Button-1>', cb2)
btCB2.place(x=210, y = 100, width=80, height=30)



window.mainloop()
```

```python
# Combobox
from tkinter import *
from calendar import month_name
from tkinter.messagebox import showinfo
from tkinter import ttk

root = Tk()
root.geometry('350x300')
root.title('Bảng giá')

# Label
label = Label(text='Please choose a month: ')
label.pack(fill=X, padx=5, pady=5)

# Create a combobox
selected_month = StringVar()
month_cb = ttk.Combobox(root, textvariable = selected_month)

# Get first 3 character of every month
month_cb['values'] = [month_name[i][0:3] for i in range(1, 13)]
# Prevent typing a value
month_cb['state'] = 'readonly'

# Place the widget
month_cb.pack(fill=X, padx=5, pady=5)

# Bind the selected value changes
def month_change(event):
    showinfo(title='Result',
             message=f'You selected {selected_month.get()}!'
             )
    
month_cb.bind('<<ComboboxSelected>>', month_change)

root.mainloop()

              
```

```python
from tkinter import *
from tkinter.messagebox import showinfo
from tkinter import ttk

def action(event):
    try:
        num1 = int(entryInput1.get())
        num2 = int(entryInput2.get())
        if thaotac.get() == '+':
            labelOutput['text']=str(num1) +' '+ thaotac.get()+' ' + str(num2) + ' = ' + str(num1+num2)
        elif thaotac.get() == '-':
            labelOutput['text']=str(num1) +' '+ thaotac.get()+' ' + str(num2) + ' = ' + str(num1-num2)
        elif thaotac.get() == '*':
            labelOutput['text']=str(num1) +' '+ thaotac.get()+' ' + str(num2) + ' = ' + str(num1*num2)
        elif thaotac.get() == '/':
            if num2 == 0:
                labelOutput['text'] = 'Lỗi chia cho 0'
            else:
                labelOutput['text']=str(num1) +' '+ thaotac.get()+' ' + str(num2) + ' = ' + str(num1/num2)
        labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
    
        
window = Tk()
window.geometry('400x250')
window.title('Toán tử')

titleInput1 = Label(window, text='Num 1:')
titleInput1.place(x=10, y= 10)
titleInput2 = Label(window, text='Num 2:')
titleInput2.place(x=10, y= 50)
titleOutput = Label(window, text='Thao tác:')
titleOutput.place(x=10, y= 90)

entryInput1 = Entry(window, text = '', width = 35)
entryInput1.place(x=100, y = 10)
entryInput2 = Entry(window, text = '', width = 35)
entryInput2.place(x=100, y = 50)
entryInput1.focus()

thaotac = StringVar()
thaotac = ttk.Combobox(window, textvariable = thaotac)
thaotac.place(x=100, y = 90)

bt = Button(window, text='Thực hiện')
bt.place(x=130, y = 140)
bt.bind('<Button-1>', action)

titleOutput = Label(window, text='Output:')
titleOutput.place(x=10, y= 200)

thaotac['values'] = ['+','-','*','/']
thaotac['state'] = 'readonly'

labelOutput = Label(window, text = '', borderwidth = 1, font=('Arial', 15))
labelOutput.place(x=100, y = 200)


window.mainloop()
```

```python
from tkinter import *
from tkinter.messagebox import showinfo
from tkinter import ttk
        
root = Tk()
root.geometry('400x250')
root.title('Toán tử')

def validate1(event):
    try:
        text_input = int(event.widget.get())
        lbError1['text'] = ''
    except:
        lbError1['text'] = 'Lỗi: Số thứ nhất chưa đúng'
        
def validate2(event):
    try:
        text_input = int(event.widget.get())
        lbError2['text'] = ''
    except:
        lbError2['text'] = 'Lỗi: Số thứ hai chưa đúng'        
        
lbNumber1 = Label(root, text='Số thứ nhất', font=('Helvetica',13))
lbNumber1.place(x=10, y=10)
txtNumber1 = Entry(root, font=('Helvetica',13))
txtNumber1.place(x=120, y=10)
lbError1 = Label(root, text='', font=('Helvetica',8,'italic'), fg='Red')
lbError1.place(x=120, y=35)
txtNumber1.bind('<FocusOut>', validate1)

lbNumber2 = Label(root, text='Số thứ hai', font=('Helvetica',13))
lbNumber2.place(x=10, y=75)
txtNumber2 = Entry(root, font=('Helvetica',13))
txtNumber2.place(x=120, y=75)
lbError2 = Label(root, text='', font=('Helvetica',8,'italic'), fg='Red')
lbError2.place(x=120, y=100)
txtNumber2.bind('<FocusOut>', validate2)

lbAction = Label(root, text='Hành động', font=('Helvetica',13))
lbAction.place(x=10, y=140)

SelectAction = StringVar()
SelectAction.set('+')
cbbAction = ttk.Combobox(root, textvariable = SelectAction)
cbbAction.place(x=120, y = 140)
cbbAction['values'] = ['+','-','*','/']
cbbAction['state'] = 'readonly'

def calculation(event):
    btCalculate.focus()
    try:
        if lbError1['text'] != '' or lbError2['text'] != '':
            lbRerult['text'] = ''
            return
        action = SelectAction.get()
        if action == '+':
            lbResult['text'] = int(txtNumber1.get()) + int(txtNumber2.get())
        elif action == '-':
            lbResult['text'] = int(txtNumber1.get()) - int(txtNumber2.get())
        elif action == '*':
            lbResult['text'] = int(txtNumber1.get()) * int(txtNumber2.get())
        elif action == '/':
            if int(txtNumber2.get()) != 0:
                lbResult['text'] = int(txtNumber1.get()) / int(txtNumber2.get())
            else:
                lbError2['text'] = 'Lỗi: Số thứ hai = 0' 
                lbResult['text']=''
    except:
        lbResult['text']=''
        
btCalculate = Button(root, text='Tính')
btCalculate.place(x=120, y = 190)
btCalculate.bind('<Button-1>', calculation)

lbResult = Label(root, text='Kết quả', font=('Helvetica',18))
lbResult.place(x=180, y = 190)

root.mainloop()
```
