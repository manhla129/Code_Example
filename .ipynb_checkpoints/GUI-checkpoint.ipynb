```python
from tkinter import *

window = Tk()
window.geometry('400x250')
window.title('Click event')

def callback(event):
    print('X, Y:', event.x, event.y)
    print('Widget:', event.widget)
    print('X-root, Y-root:', event.x_root, event.y_root)
    
def click(event):
    print('You click on me seccond time')
    
def enter(event):
    print('You enter me')
    event.widget['bg'] = 'blue'
    
def leave(event):
    print('You leave me')
    event.widget['bg'] = 'red'
    
def entered(event):
    print('Oh, you press enter on me')
    
def key(event):
    print('Key pressed: ', event.char)

    
button = Button(window, text='Confirm')
button2 = Button(window, text='Exit')
# frame = Frame(window, width=400, height=250)
button.bind('<Button-1>', callback)
button.bind('<Button-1>', click, add='+')
button.bind('<Return>', entered)
button.bind('<Key>', key)
# button.bind('<Enter>', enter)
# button.bind('<Leave>', leave)
button.focus()
button.pack()
button2.pack()

window.mainloop()
```

```python
from tkinter import *

window = Tk()
window.geometry('400x250')
window.title('Click event')

def gt(event):
    number = txtNumber.get()
    result = 1
    for i in range(2, int(number)+1):
        result *= i
    lbMessage['text'] = result
    
txtNumber = Entry(window, text='')
txtNumber.pack()

lbMessage = Label(window, text='Out Message')
lbMessage.pack()

button = Button(window, text='Giai thừa')
button.bind('<Button-1>', gt)
button.pack()

window.mainloop()
```

```python
from tkinter import *

def gt(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            result = 1
            for i in range(2, int(number)+1):
                result *= i
            labelOutput['text'] = result
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
    
def lt(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            number = entryInput.get()
            labelOutput['text'] = (int(number))**2
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
def cb2(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            number = entryInput.get()
            labelOutput['text'] = (int(number))**0.5
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
        
window = Tk()
window.geometry('400x250')
window.title('Click event')

titleInput = Label(window, text='Input:')
titleInput.place(x=10, y= 10)
titleOutput = Label(window, text='Output:')
titleOutput.place(x=10, y= 50)

entryInput = Entry(window, text = '', width = 35)
entryInput.place(x=100, y = 10)
labelOutput = Label(window, text = '', borderwidth =1)
labelOutput.place(x=100, y = 50)
entryInput.focus()

btGT = Button(window, text='Giai thừa')
btGT.bind('<Button-1>', gt)
btGT.place(x=30, y = 100, width=70, height=30)

btLT = Button(window, text='Lũy thừa')
btLT.bind('<Button-1>', lt)
btLT.place(x=120, y = 100, width=70, height=30)

btCB2 = Button(window, text='Căn bậc 2')
btCB2.bind('<Button-1>', cb2)
btCB2.place(x=210, y = 100, width=80, height=30)



window.mainloop()
```

```python
# Radiobutton
from tkinter import *

def sel():
    selection = 'You selected the option: ' + var.get()
    label.config(text=selection)
#     label['text']  = selection

root = Tk()
root.geometry('350x300')

# var = IntVar()
var = StringVar()
var.set('Other')
gender = LabelFrame(root, width=450, height=200, text='Gender')

rd1 = Radiobutton(gender, text='Female', variable=var, value='Male', command=sel)
rd1.pack(anchor = W)
rd2 = Radiobutton(gender, text='Male', variable=var, value='Female', command=sel)
rd2.pack(anchor = W)
rd3 = Radiobutton(gender, text='Other', variable=var, value='Other', command=sel)
rd3.pack(anchor = W)

gender.pack(expand=True)

label  = Label(root)
label.pack()
root.mainloop()
```

```python
# Checkbutton
from tkinter import *

def sel():
    music = CheckVar1.get()
    video = CheckVar2.get()
    label['text'] = 'music: ' + str(music) + ' video: ' + str(video)

root = Tk()
CheckVar1 = IntVar()
CheckVar2 = IntVar()

C1 = Checkbutton(root, text='Music',variable=CheckVar1,
                onvalue=1, offvalue=0, height=5, width=20,
                command=sel)
C2 = Checkbutton(root, text='Video',variable=CheckVar2,
                onvalue=1, offvalue=0, height=5, width=20,
                command=sel)
C1.pack()
C2.pack()
label  = Label(root, text='')
label.pack()
root.mainloop()
```

```python
# Checkbutton
from tkinter import *

def sel():
    bunbo = int(CheckVar1.get())
    buncha = int(CheckVar2.get())
    bundau = int(CheckVar3.get())
    trada = int(CheckVar4.get())
    thanhtoan = int(CheckVar5.get())/100
    label['text'] = str((bunbo + buncha + bundau + trada)*thanhtoan) + 'k'

root = Tk()
root.geometry('350x300')
root.title('Bảng giá')
menu = LabelFrame(root, width=450, height=200, text='Menu')
menu.place(x=10, y=10)

CheckVar1 = IntVar()
CheckVar2 = IntVar()
CheckVar3 = IntVar()
CheckVar4 = IntVar()

C1 = Checkbutton(menu, text='Bún bò (55k)',variable=CheckVar1,
                onvalue=55, offvalue=0, height=3, width=15,
                command=sel)
C2 = Checkbutton(menu, text='Bún chả (40k)',variable=CheckVar2,
                onvalue=40, offvalue=0, height=3, width=15,
                command=sel)
C3 = Checkbutton(menu, text='Bún đậu (50k)',variable=CheckVar3,
                onvalue=50, offvalue=0, height=3, width=15,
                command=sel)
C4 = Checkbutton(menu, text='Trà đá (10k)',variable=CheckVar4,
                onvalue=10, offvalue=0, height=3, width=15,
                command=sel)
C1.pack(anchor = W)
C2.pack(anchor = W)
C3.pack(anchor = W)
C4.pack(anchor = W)

thanhtoan = LabelFrame(root, width=150, height=100, text='Phương thức thanh toán')
thanhtoan.place(x=170, y=10)

CheckVar5 = IntVar()
CheckVar5.set(95)

rd1 = Radiobutton(thanhtoan, text='Tiền mặt (-5%)',variable=CheckVar5,
                value = 95,
                command=sel)
rd2 = Radiobutton(thanhtoan, text='Chuyển khoản (+5%)',variable=CheckVar5,
                value = 105,
                command=sel)
rd1.pack(anchor = W)
rd2.pack(anchor = W)

tongtien = LabelFrame(root, width=150, height=10, text='Tổng tiền')
tongtien.place(x=170, y=100)


label  = Label(tongtien, text='', font=('Arial',16))
label.pack(anchor = W)
root.mainloop()
```
