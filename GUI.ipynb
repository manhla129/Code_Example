```python
from tkinter import *

window = Tk()
window.geometry('400x250')
window.title('Click event')

def callback(event):
    print('X, Y:', event.x, event.y)
    print('Widget:', event.widget)
    print('X-root, Y-root:', event.x_root, event.y_root)
    
def click(event):
    print('You click on me seccond time')
    
def enter(event):
    print('You enter me')
    event.widget['bg'] = 'blue'
    
def leave(event):
    print('You leave me')
    event.widget['bg'] = 'red'
    
def entered(event):
    print('Oh, you press enter on me')
    
def key(event):
    print('Key pressed: ', event.char)

    
button = Button(window, text='Confirm')
button2 = Button(window, text='Exit')
# frame = Frame(window, width=400, height=250)
button.bind('<Button-1>', callback)
button.bind('<Button-1>', click, add='+')
button.bind('<Return>', entered)
button.bind('<Key>', key)
# button.bind('<Enter>', enter)
# button.bind('<Leave>', leave)
button.focus()
button.pack()
button2.pack()

window.mainloop()
```

```python
from tkinter import *

window = Tk()
window.geometry('400x250')
window.title('Click event')

def gt(event):
    number = txtNumber.get()
    result = 1
    for i in range(2, int(number)+1):
        result *= i
    lbMessage['text'] = result
    
txtNumber = Entry(window, text='')
txtNumber.pack()

lbMessage = Label(window, text='Out Message')
lbMessage.pack()

button = Button(window, text='Giai thừa')
button.bind('<Button-1>', gt)
button.pack()

window.mainloop()
```

```python
from tkinter import *

def gt(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            result = 1
            for i in range(2, int(number)+1):
                result *= i
            labelOutput['text'] = result
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
    
def lt(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            number = entryInput.get()
            labelOutput['text'] = (int(number))**2
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
def cb2(event):
    try:
        number = entryInput.get()
        if int(number) < 0:
            raise ValueError('Lỗi nhập dữ liệu')
        else:
            number = entryInput.get()
            labelOutput['text'] = (int(number))**0.5
            labelOutput['fg'] = 'green'
    except:
        labelOutput['text'] = 'Lỗi đầu vào'
        labelOutput['fg'] = 'red'
        
        
window = Tk()
window.geometry('400x250')
window.title('Click event')

titleInput = Label(window, text='Input:')
titleInput.place(x=10, y= 10)
titleOutput = Label(window, text='Output:')
titleOutput.place(x=10, y= 50)

entryInput = Entry(window, text = '', width = 35)
entryInput.place(x=100, y = 10)
labelOutput = Label(window, text = '', borderwidth =1)
labelOutput.place(x=100, y = 50)
entryInput.focus()

btGT = Button(window, text='Giai thừa')
btGT.bind('<Button-1>', gt)
btGT.place(x=30, y = 100, width=70, height=30)

btLT = Button(window, text='Lũy thừa')
btLT.bind('<Button-1>', lt)
btLT.place(x=120, y = 100, width=70, height=30)

btCB2 = Button(window, text='Căn bậc 2')
btCB2.bind('<Button-1>', cb2)
btCB2.place(x=210, y = 100, width=80, height=30)



window.mainloop()
```
